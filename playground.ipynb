```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine, text

load_dotenv()

DB_URL = f"postgresql+{os.getenv('DATABASE_URL')}:{os.getenv('POSTGRES_PASSWORD')}@{os.getenv('POSTGRES_HOST')}:{os.getenv('POSTGRES_PORT', '5432')}/{os.getenv('POSTGRES_DB')}"

def delete_customer_data_single_table(customer_id: str):
    """Delete all records for specific customer from single table"""
    if not customer_id:
        print("❌ Error: Customer ID is required")
        return False
    print(f"Current DB_URL: '{DB_URL}'") 
    engine = create_engine(DB_URL)

    try:
        with engine.connect() as conn:
            # Single query to delete everything
            result = conn.execute(
                text("DELETE FROM email_claim WHERE customer_id = :cust_id"),
                {"cust_id": customer_id}
            )
            conn.commit()
            print(f"🗑️ Deleted {result.rowcount} records for customerid")
    except Exception as e:
        print(f"failed: {e}")

def view_email_claim_sample():
    engine = create_engine(DB_URL)
    try:
        with engine.connect() as conn:
            result = conn.execute(text("SELECT * FROM email_claim LIMIT 10"))
            rows = result.fetchall()
            print(rows)
    except Exception as e:
        print("Error:", e)


if __name__ == "__main__":
    delete_customer_data_single_table("CUS000002")
    view_email_claim_sample()
```
